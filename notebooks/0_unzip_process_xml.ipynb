```python
if icorps2:
    # Load the Excel file
    icorps = pd.read_excel('../Data/icorps_coded.xlsx')
    # Columns to process
    columns_to_clean = ['teams', 'hub', 'site', 'node']

    # Process each column
    for col in columns_to_clean:
        print(f"Processing column: {col}")
        
        # Replace NaN values with 0
        icorps[col] = icorps[col].apply(lambda x: 0 if pd.isnull(x) else x)
        
        # If the value is a float, convert it to an integer directly (to remove .0)
        icorps[col] = icorps[col].apply(lambda x: int(x) if isinstance(x, float) else x)
        
        # Convert to string to handle potential non-numeric values
        icorps[col] = icorps[col].astype(str)
        
        # Remove all non-numeric characters
        icorps[col] = icorps[col].str.replace(r'\D', '', regex=True)
        
        # Convert back to integers (defaulting empty strings to 0)
        icorps[col] = icorps[col].apply(lambda x: int(x) if x.isdigit() else 0)
        icorps['AwardID'] = icorps['AwardID'].astype(str)

    # Final conversion to ensure proper integer type
    icorps[columns_to_clean] = icorps[columns_to_clean].astype(int)
    icorps[['AwardID','teams', 'hub', 'site', 'node']].to_csv(output_csv / 'final' / 'icorps_final.csv', index=False)
    icorps[['AwardID','teams', 'hub', 'site', 'node']].to_pickle(output_pkl / 'final' / 'icorps_final.pkl')


icorps
```
