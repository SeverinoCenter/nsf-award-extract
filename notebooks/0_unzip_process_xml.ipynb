```python
#Create important grant features for use later. 
grants=utils.create_grant_features(grants)
```

```python
#Feature Engineering
# Add a new column 'Is_SBIR' to the grants DataFrame
grants['sbir'] = grants['AwardTitle'].str.contains('SBIR ', case=False, na=False).astype(int)
grants['sbir_1'] = grants['AwardTitle'].str.contains('SBIR Phase I:', case=False, na=False).astype(int)
grants['sbir_2'] = grants['AwardTitle'].str.contains('SBIR Phase II:', case=False, na=False).astype(int)
grants['sttr'] = grants['AwardTitle'].str.contains('STTR ', case=False, na=False).astype(int)
grants['sttr_1'] = grants['AwardTitle'].str.contains('STTR Phase I:', case=False, na=False).astype(int)
grants['sttr_2'] = grants['AwardTitle'].str.contains('STTR Phase II:', case=False, na=False).astype(int)

```

```python
# Sum each feature column and store results in a dictionary
feature_sums = {
    'total_sbir': grants['sbir'].sum(),
    'total_sbir_1': grants['sbir_1'].sum(),
    'total_sbir_2': grants['sbir_2'].sum(),
    'total_sttr': grants['sttr'].sum(),
    'total_sttr_1': grants['sttr_1'].sum(),
    'total_sttr_2': grants['sttr_2'].sum()
}

# Convert to a DataFrame for easy visualization
feature_sums_df = pd.DataFrame(list(feature_sums.items()), columns=['Feature', 'Total'])

# Display the results
feature_sums_df
```
