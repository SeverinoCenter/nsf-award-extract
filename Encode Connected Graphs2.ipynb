```python
#select only AwardID and Investigator columns
ic=[x for x in xml_df.columns if 'Investigator' in x]
ic=['AwardID']+ic
xml_df = xml_df[ic]
```

```python
# Define the number of investigators
num_investigators = 3

# Loop through each investigator number and rename columns
for i in range(1, num_investigators + 1):
    for col in xml_df.columns:
        if f'Investigator{i}_' in col:
            new_col_name = col.replace(f'Investigator{i}_', f'Investigator_') + str(i)
            xml_df.rename(columns={col: new_col_name}, inplace=True)
    # Add a new column for InvestigatorNumber with the value i
    xml_df[f'InvestigatorNumber{i}'] = i
```

```python
# Convert wide to long format
investigator_columns = [
    'FirstName', 'LastName', 'MiddleInitial', 'Suffix', 
    'FullName', 'Email', 'NSFID', 'StartDate', 
    'EndDate', 'RoleCode'
]
stubnames = [f'Investigator_{col}' for col in investigator_columns] + ['InvestigatorNumber']
long_df = pd.wide_to_long(xml_df,
                          stubnames=stubnames,
                          i=[col for col in xml_df.columns if not any(name in col for name in stubnames)],
                          j='InvestigatorNumber').reset_index()
```

```python
# Convert wide to long format
long_df = pd.wide_to_long(xml_df,
                          stubnames=[
                              'Investigator_FirstName', 'Investigator_LastName', 'Investigator_MiddleInitial', 
                              'Investigator_Suffix', 'Investigator_FullName', 'Investigator_Email', 'Investigator_NSFID', 
                              'Investigator_StartDate', 'Investigator_EndDate', 'Investigator_RoleCode', 'InvestigatorNumber'
                          ],
                          i=[col for col in xml_df.columns if not col.startswith('Investigator_')],
                          j='InvestigatorNumber',
                          sep='').reset_index()

# Print the resulting DataFrame
print("Long DataFrame:")
long_df.head()
```
