```python
icorps_grants = df[df['title'].str.contains('I-Corps', na=False)]
sbir_sttr_grants = df[df['title'].str.contains('SBIR|STTR', na=False)]
```

```python
print("I-Corps Grants:\n")
icorps_grants.head()
```

```python
print("SBIR/STTR Grants:\n")
sbir_sttr_grants.head()
```

```python
merged_df = icorps_grants.merge(sbir_sttr_grants, on='poName', suffixes=('_icorps', '_sbir'))

merged_df['same_idea'] = merged_df.apply(lambda x: x['abstractText_icorps'] == x['abstractText_sbir'], axis=1)

print("Merged DataFrame with Idea Comparison:\n")
merged_df[['poName', 'abstractText_icorps', 'abstractText_sbir', 'same_idea']]
```

```python
summary = merged_df['same_idea'].value_counts()

print("Summary of Idea Transitions:\n", summary)
```

```python
ruth_shuman_records = merged_df[merged_df['poName'] == 'Ruth Shuman']

print("All Records for Ruth Shuman:\n")
ruth_shuman_records
```

```python
ruth_shuman_records = df[df['poName'] == 'Ruth Shuman']

print("All Records for Ruth Shuman:\n")
ruth_shuman_records
```

```python
changed_ideas_df = merged_df[merged_df['same_idea'] == True]

print("Sample of Not Changed Ideas:\n")
changed_ideas_df[['poName', 'abstractText_icorps', 'abstractText_sbir']].head(10)
```

```python
changed_ideas_df = merged_df[merged_df['same_idea'] == False]

print("Sample of Changed Ideas:\n")
changed_ideas_df[['poName', 'abstractText_icorps', 'abstractText_sbir']].head(10)
```

```python
def categorize_change(row):
    if 'improve' in row['abstractText_sbir'].lower() or 'enhance' in row['abstractText_sbir'].lower():
        return 'Improvement'
    elif 'new' in row['abstractText_sbir'].lower() or 'novel' in row['abstractText_sbir'].lower():
        return 'New Project'
    else:
        return 'Other'

changed_ideas_df['change_category'] = changed_ideas_df.apply(categorize_change, axis=1)

print("Categorized Changes:\n")
changed_ideas_df[['poName', 'abstractText_icorps', 'abstractText_sbir', 'change_category']].head(10)
```

```python
changed_ideas_df['change_category'] = changed_ideas_df.apply(categorize_change, axis=1)

category_counts = changed_ideas_df['change_category'].value_counts()

print("Number of rows in each category:\n", category_counts)
```

```python
changed_ideas_df.to_csv('icorps_to_sbir_categorized.csv', index=False)
```
