```python
import pandas as pd
```

```python
xml_df = pd.read_csv('../Data/cleaned_xml_combined_08112024.csv')
```

```python
print("Original DataFrame:")
xml_df.head()
```

```python
xml_df.columns
```

```python
investigator_columns = [
    'Investigator1_FullName', 'Investigator1_Email', 'Investigator1_NSFID', 'Investigator1_StartDate', 'Investigator1_EndDate', 'Investigator1_RoleCode',
    'Investigator2_FullName', 'Investigator2_Email', 'Investigator2_NSFID', 'Investigator2_StartDate', 'Investigator2_EndDate', 'Investigator2_RoleCode',
    'Investigator3_FullName', 'Investigator3_Email', 'Investigator3_NSFID', 'Investigator3_StartDate', 'Investigator3_EndDate', 'Investigator3_RoleCode'
]

# melt operation to convert wide format to long format
melted_df = pd.melt(xml_df, 
                    id_vars=[col for col in xml_df.columns if col not in investigator_columns], 
                    value_vars=investigator_columns, 
                    var_name='Investigator_Info', 
                    value_name='InvestigatorValue')

print("Melted DataFrame:")
melted_df.head()
```

```python
# Extract the Investigator ID and the attribute
melted_df['InvestigatorID'] = melted_df['Investigator_Info'].str.extract(r'(\d)').astype(int)
melted_df['Attribute'] = melted_df['Investigator_Info'].str.extract(r'_(.*)')

print("Columns after extracting InvestigatorID and Attribute:", melted_df.columns)

print("DataFrame after extracting InvestigatorID and Attribute:")
melted_df.head()
```

```python
required_columns = [
    'AwardID', 'AwardTitle', 'AwardEffectiveDate', 'AwardExpirationDate', 'Institution_Name', 'AbstractNarration'
]

# Pivot the table to get a cleaner long format using groupby and unstack
pivot_df = melted_df.pivot_table(index=['AwardID', 'InvestigatorID'], columns='Attribute', values='InvestigatorValue', aggfunc='first').reset_index()

pivot_df.columns = [col.replace('_', '') for col in pivot_df.columns]

pivot_df = pivot_df.merge(xml_df[required_columns].drop_duplicates(), on='AwardID', how='left')

print("Columns in pivoted DataFrame:", pivot_df.columns)

print("Pivoted DataFrame:")
pivot_df.head()
```

```python
cleaned_df = pivot_df.dropna(subset=['FullName'])

# unique PersonID by combining FullName and Email
cleaned_df['PersonID'] = cleaned_df['FullName'] + '_' + cleaned_df['Email']

cleaned_df['PERSON-Icorps'] = False
cleaned_df['PERSON-sbir'] = False
cleaned_df['PERSON-research'] = False
cleaned_df['PERSON-Sttr'] = False
```

```python
icorps_count = cleaned_df['AwardTitle'].str.contains('I-Corps', case=False, na=False).sum()
sbir_count = cleaned_df['AwardTitle'].str.contains('SBIR', case=False, na=False).sum()
research_count = cleaned_df['AwardTitle'].str.contains('Research', case=False, na=False).sum()
sttr_count = cleaned_df['AwardTitle'].str.contains('STTR', case=False, na=False).sum()

total_entries = len(cleaned_df)

icorps_not_count = total_entries - icorps_count
sbir_not_count = total_entries - sbir_count
research_not_count = total_entries - research_count
sttr_not_count = total_entries - sttr_count

print(f"Number of entries with 'I-Corps': {icorps_count}")
print(f"Number of entries without 'I-Corps': {icorps_not_count}")
print(f"Number of entries with 'SBIR': {sbir_count}")
print(f"Number of entries without 'SBIR': {sbir_not_count}")
print(f"Number of entries with 'Research': {research_count}")
print(f"Number of entries without 'Research': {research_not_count}")
print(f"Number of entries with 'STTR': {sttr_count}")
print(f"Number of entries without 'STTR': {sttr_not_count}")
```

```python
# Sample logic to classify grants (this should be adapted to your specific logic)
cleaned_df.loc[cleaned_df['AwardTitle'].str.contains('I-Corps', case=False, na=False), 'PERSON-Icorps'] = True
cleaned_df.loc[cleaned_df['AwardTitle'].str.contains('SBIR', case=False, na=False), 'PERSON-sbir'] = True
cleaned_df.loc[cleaned_df['AwardTitle'].str.contains('Research', case=False, na=False), 'PERSON-research'] = True
cleaned_df.loc[cleaned_df['AwardTitle'].str.contains('STTR', case=False, na=False), 'PERSON-Sttr'] = True

# Display the transformed and cleaned DataFrame
print("Transformed and Cleaned DataFrame:")
cleaned_df.head()
```

```python
cleaned_df['AwardEffectiveDate'] = pd.to_datetime(cleaned_df['AwardEffectiveDate'])
cleaned_df = cleaned_df.sort_values(by=['PersonID', 'AwardEffectiveDate'])

# Function to identify transitions
def identify_transitions(df):
    transitions = {
        'Research_to_Icorps': 0,
        'Research_to_Sbir': 0,
        'Research_to_Sttr': 0,
        'Icorps_to_Research': 0,
        'Icorps_to_Sbir': 0,
        'Icorps_to_Sttr': 0,
        'Sbir_to_Research': 0,
        'Sbir_to_Icorps': 0,
        'Sbir_to_Sttr': 0,
        'Sttr_to_Research': 0,
        'Sttr_to_Icorps': 0,
        'Sttr_to_Sbir': 0
    }
    
    previous_grant = None
    
    for index, row in df.iterrows():
        current_grant = None
        if row['PERSON-research']:
            current_grant = 'Research'
        elif row['PERSON-Icorps']:
            current_grant = 'Icorps'
        elif row['PERSON-sbir']:
            current_grant = 'Sbir'
        elif row['PERSON-Sttr']:
            current_grant = 'Sttr'
        
        if previous_grant and current_grant and previous_grant != current_grant:
            transitions[f'{previous_grant}_to_{current_grant}'] += 1
        
        previous_grant = current_grant
    
    return transitions

# Group by PersonID and apply the identify_transitions function
all_transitions = cleaned_df.groupby('PersonID').apply(identify_transitions).reset_index(drop=True)

# Aggregate transitions
total_transitions = all_transitions.apply(pd.Series).sum()

print("Transitions between different grant types:")
print(total_transitions)
```
