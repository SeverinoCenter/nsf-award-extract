```python
df=pd.read_csv(path/'allNSF_2024_06_17.csv')
#drop first column from df
df=df.drop(df.columns[0], axis=1)
#trim th ewhitespace from the id field using lambda function
df['id'] = df['id'].apply(lambda x: str(x).strip())

```

```python
#remove duplicates from df
df2 = df.drop_duplicates()
df2['id_hash'] = df2['id'].apply(lambda x: hash(x))
#move 'id to the first column of the dataframe df2
cols = list(df2.columns)
cols.insert(0, cols.pop(cols.index('id_hash')))
cols.insert(0, cols.pop(cols.index('id')))
df2.sort_values(by='id_hash', inplace=True)
df2 = df2[cols]
#drop duplicates by id_hash

#show duplicates by id_hash
duplicates=df2[df2.duplicated(subset='id_hash', keep=False)]
duplicates
```

```python
duplicates.reset_index(drop=True, inplace=True)
duplicates.to_csv(path/'duplicates.csv', index=False)
```
