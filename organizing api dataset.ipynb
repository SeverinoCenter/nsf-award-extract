```python
import pandas as pd

api_data_path = "../Data/allNSF_2024_06_17_no_duplicates.csv"  
api_df = pd.read_csv(api_data_path)

combined_data_path = "../Data/output_data_20240710_151858.csv"  
combined_df = pd.read_csv(combined_data_path)
```

```python
api_df.head()
```

```python
# Rename columns in api_df to match combined_df
api_df.rename(columns={
    'id': 'AwardID',
    'title': 'AwardTitle',
    'agency': 'Agency',
    'startDate': 'AwardEffectiveDate',
    'expDate': 'AwardExpirationDate',
    'estimatedTotalAmt': 'AwardTotalIntnAmount',
    'fundsObligatedAmt': 'AwardAmount',
    'primaryProgram': 'AwardInstrument',
    'awardee': 'Organization_Code',
    'coPDPI': 'Directorate_Abbreviation',
    'fundProgramName': 'Directorate_LongName',
    'perfLocation': 'Division_Abbreviation',
    'projectOutComesReport': 'Division_LongName',
    'poName': 'ProgramOfficer_Name',
    'poEmail': 'ProgramOfficer_Email',
    'poPhone': 'ProgramOfficer_Phone',
    'abstractText': 'AbstractNarration',
    'date': 'MinAmdLetterDate',  # Assuming 'date' here is similar
    'publicationResearch': 'MaxAmdLetterDate',  # Placeholder if available
    'transType': 'TRAN_TYPE',
    'cfdaNumber': 'CFDA_NUM',
    'fundAgencyCode': 'FUND_AGCY_CODE',
    'awardAgencyCode': 'AWDG_AGCY_CODE',
    'pdPIName': 'Investigator1_Name',
    'piEmail': 'Investigator1_Email',
    'awardeeName': 'Institution_Name',
    'awardeeCity': 'Institution_City',
    'awardeeStateCode': 'Institution_State',
    'awardeeZipCode': 'Institution_Zip',
    'awardeeCountryCode': 'Institution_Country',
    'piPhone': 'Institution_Phone',  # Assuming 'piPhone' corresponds to the phone
    'perfName': 'Performance_Institution_Name',
    'perfCity': 'Performance_Institution_City',
    'perfStateCode': 'Performance_Institution_State',
    'perfZipCode': 'Performance_Institution_Zip',
    'perfCountryCode': 'Performance_Institution_Country',
}, inplace=True)
```

```python
# Ensure startDate is treated as a string (if not already)
api_df['AwardEffectiveDate'] = api_df['AwardEffectiveDate'].astype(str)

# Convert startDate to datetime and extract the year
api_df['AwardEffectiveDate'] = pd.to_datetime(api_df['AwardEffectiveDate'], errors='coerce')
api_df['Year'] = api_df['AwardEffectiveDate'].dt.year

# Add columns that are in combined_df but not in api_df, with NaN values
for col in combined_df.columns:
    if col not in api_df.columns:
        api_df[col] = pd.NA

api_df.head()
```

```python
updated_api_data_path = "../Data/allNSF_updated_with_year.csv"
api_df.to_csv(updated_api_data_path, index=False)

print(f'Updated API data with Year column saved to {updated_api_data_path}')
```
