```python
xml_df['AwardID'] = xml_df['AwardID'].astype(str)
api_df['AwardID'] = api_df['AwardID'].astype(str)

# Find new records in api_df that are not in combined_df
new_records = api_df[~api_df['AwardID'].isin(xml_df['AwardID'])]

print(f'New records count: {len(new_records)}')

# Find common records
common_awards = api_df[api_df['AwardID'].isin(xml_df['AwardID'])]

# Merge on AwardID to align records
merged_df = common_awards.merge(xml_df, on='AwardID', suffixes=('_api', '_combined'))

# Compare columns to find differences
diff_records = []

for index, row in merged_df.iterrows():
    differences = {'AwardID': row['AwardID']}
    for column in xml_df.columns:
        if column != 'AwardID' and row[column + '_api'] != row[column + '_combined']:
            differences[column + '_api'] = row[column + '_api']
            differences[column + '_combined'] = row[column + '_combined']
    if len(differences) > 1:  # means there are differences other than AwardID
        diff_records.append(differences)

diff_records_df = pd.DataFrame(diff_records)

print(f'Total records with differences: {len(diff_records_df)}')
```

```python
from datetime import datetime
# Get the current date and time
now = datetime.now()
date_time_str = now.strftime("%Y%m%d_%H%M%S")

# Save new records to CSV
new_records_path = f'new_records_{date_time_str}.csv'
new_records.to_csv(new_records_path, index=False)

# Save differences to CSV
diff_records_path = f'diff_records_{date_time_str}.csv'
diff_records_df.to_csv(diff_records_path, index=False)

print(f'New records saved to {new_records_path}')
print(f'Differences saved to {diff_records_path}')
```

## examine only those values for awardIDs that are in both of the datasets. (1) records which are in both the API and the XML (2) missing value analysis

```python
xml_df['AwardID'] = xml_df['AwardID'].astype(str)
api_df['AwardID'] = api_df['AwardID'].astype(str)

common_award_ids = set(xml_df['AwardID']).intersection(set(api_df['AwardID']))
common_xml_records = xml_df[xml_df['AwardID'].isin(common_award_ids)]
common_api_records = api_df[api_df['AwardID'].isin(common_award_ids)]

print(f"Number of common records: {len(common_award_ids)}")
```

```python
common_columns = set(common_xml_records.columns).intersection(set(common_api_records.columns))
common_xml_records = common_xml_records[common_columns]
common_api_records = common_api_records[common_columns]
```

```python
missing_values_common_xml = common_xml_records.isnull().sum()
print("Missing value analysis for common records in XML DataFrame:")
missing_values_common_xml
```

```python
missing_values_common_api = common_api_records.isnull().sum()
print("Missing value analysis for common records in API DataFrame:")
missing_values_common_api
```

```python
common_xml_records = common_xml_records.add_suffix('_XML')
common_api_records = common_api_records.add_suffix('_API')

if 'Year_XML' not in common_xml_records.columns:
    common_xml_records['Year_XML'] = xml_df['Year']
if 'Year_API' not in common_api_records.columns:
    common_api_records['Year_API'] = api_df['Year']

common_xml_missing_by_year = common_xml_records.groupby('Year_XML').apply(lambda x: x.isnull().sum())
common_xml_missing_by_year['Total Records_XML'] = common_xml_records.groupby('Year_XML').size()

common_api_missing_by_year = common_api_records.groupby('Year_API').apply(lambda x: x.isnull().sum())
common_api_missing_by_year['Total Records_API'] = common_api_records.groupby('Year_API').size()

merged_missing_by_year_common = pd.concat([common_xml_missing_by_year, common_api_missing_by_year], axis=1)

xml_columns = [col for col in merged_missing_by_year_common.columns if col.endswith('_XML')]
api_columns = [col.replace('_XML', '_API') for col in xml_columns]
ordered_columns = [val for pair in zip(xml_columns, api_columns) for val in pair] + ['Total Records_XML', 'Total Records_API']

merged_missing_by_year_common = merged_missing_by_year_common[ordered_columns]

print("Comparison of missing values by year for common records in both XML and API DataFrames:")
merged_missing_by_year_common
```
