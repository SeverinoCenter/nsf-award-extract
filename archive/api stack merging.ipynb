```python
import pandas as pd
```

```python
df1 = pd.read_csv('../Data/Merging new/DEO_24_06_24.csv')
df2 = pd.read_csv('../Data/Merging new/DOD_24_06_24.csv')
df3 = pd.read_csv('../Data/Merging new/Award_Abstract_NSF_2024_06_21.csv')
df4 = pd.read_csv('../Data/Merging new/I-Corps_Sites_2024_06_21.csv')
df5 = pd.read_csv('../Data/Merging new/NSF Org_24_06_24.csv')
df6 = pd.read_csv('../Data/Merging new/NSF_new_2024_06_24.csv')
df7 = pd.read_csv('../Data/Merging new/NSF_new_api2_2024_06_24.csv')
df8 = pd.read_csv('../Data/Merging new/PROJECT OUTCOMES REPORT_24_06_24.csv')
df9 = pd.read_csv('../Data/Merging new/allNSF.csv')
```

```python
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
for i, df in enumerate(dataframes, 1):
    print(f"Size of df{i}: {df.shape}")
```

```python


merged_df = pd.concat(dataframes, axis=0)

merged_df.reset_index(drop=True, inplace=True)

merged_df.to_csv('../Data/merged_dataframe_api_15July24.csv', index=False)

print("Shape of the merged dataframe:", merged_df.shape)
```

```python
total_rows_individual = sum(df.shape[0] for df in dataframes)
total_rows_merged = merged_df.shape[0]

if total_rows_individual == total_rows_merged:
    print("\nMerge is correct. Total rows match.")
else:
    print("\nMerge might be incorrect. Total rows do not match.")

print(f"Total rows in individual dataframes: {total_rows_individual}")
print(f"Total rows in merged dataframe: {total_rows_merged}")
```
