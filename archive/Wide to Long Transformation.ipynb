```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
#import Path
from pathlib import Path
#base = Path('/Users/jasonkuruzovich/Library/CloudStorage/Box-Box/Documents/0_research/eship_vitor/I_CORPS/data/2024_617/NSF-Downloads-8july2024/')


#xml_df = pd.read_csv(base/'cleaned_xml_combined_08112024.csv')
xml_df = pd.read_csv('../Data/NSF_grants_24.csv')
```

```python
print("Original DataFrame:")
xml_df.head()
```

```python
num_investigators = 3

# Loop through each investigator number and rename columns
for i in range(1, num_investigators + 1):
    for col in xml_df.columns:
        if f'Investigator{i}_' in col:
            new_col_name = col.replace(f'Investigator{i}_', f'Investigator_') + str(i)
            xml_df.rename(columns={col: new_col_name}, inplace=True)
```

```python
print("Columns after renaming:")
xml_df.columns
```

```python
# List of investigator columns
investigator_columns = [col for col in xml_df.columns if col.startswith('Investigator_')]

# a. Calculate the number of NA in investigator lines by year
na_investigator_by_year = xml_df[investigator_columns].isna().groupby(xml_df['Year']).sum()
# Aggregate the NA values for all investigator columns per year
na_investigator_by_year_agg = na_investigator_by_year.sum(axis=1)

# Exclude the "Historical" category and focus on a smaller time range
na_investigator_by_year_agg = na_investigator_by_year_agg[na_investigator_by_year_agg.index != 'Historical']
na_investigator_by_year_agg = na_investigator_by_year_agg[pd.to_numeric(na_investigator_by_year_agg.index, errors='coerce') >= 2000]

# b. Calculate the total NA values in all columns by year
na_total_by_year = xml_df.isna().groupby(xml_df['Year']).sum()
# Aggregate the NA values for all columns per year
na_total_by_year_agg = na_total_by_year.sum(axis=1)

# Exclude the "Historical" category
na_total_by_year_agg = na_total_by_year_agg[na_total_by_year_agg.index != 'Historical']
na_total_by_year_agg = na_total_by_year_agg[pd.to_numeric(na_total_by_year_agg.index, errors='coerce') >= 2000]

# Convert index to numeric for plotting
na_investigator_by_year_agg.index = pd.to_numeric(na_investigator_by_year_agg.index, errors='coerce')
na_total_by_year_agg.index = pd.to_numeric(na_total_by_year_agg.index, errors='coerce')

# Drop any NaN indices that may have resulted from conversion
na_investigator_by_year_agg.dropna(inplace=True)
na_total_by_year_agg.dropna(inplace=True)
```

```python
print("Number of NA values in investigator lines by year (2000 and later):")
na_investigator_by_year_agg
```

```python
print("\nTotal number of NA values in all columns by year (2000 and later):")
na_total_by_year_agg
```

```python
# Save the results to CSV files
na_investigator_by_year_agg.to_csv('../data/na_investigator_by_year.csv', header=['Total_NA_Investigator'])
na_total_by_year_agg.to_csv('../data/na_total_by_year.csv', header=['Total_NA_All_Columns'])
```

```python
# Plot for investigator columns
plt.figure(figsize=(12, 6))
bars = plt.bar(na_investigator_by_year_agg.index, na_investigator_by_year_agg, color='blue')
plt.title('Total NA values in Investigator Columns by Year')
plt.xlabel('Year')
plt.ylabel('Total NA values')
plt.xticks(rotation=45)

# Add labels to bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
# plt.savefig('/mnt/data/na_investigator_by_year.png')
plt.show()
```

```python
# Plot for all columns
plt.figure(figsize=(10, 6))
na_total_by_year_agg.plot(kind='bar', color='green')
plt.title('Total NA values in All Columns by Year')
plt.xlabel('Year')
plt.ylabel('Total NA values')
plt.xticks(rotation=45)
plt.tight_layout()
#plt.savefig('/mnt/data/na_total_by_year.png')
plt.show()
```

```python
# Convert wide to long format
investigator_columns = [
    'FirstName', 'LastName', 'MiddleInitial', 'Suffix', 
    'FullName', 'Email', 'NSFID', 'StartDate', 
    'EndDate', 'RoleCode'
]

# Create the stubnames list for wide_to_long conversion
stubnames = [f'Investigator_{col}' for col in investigator_columns]

# Perform the wide to long conversion
long_df = pd.wide_to_long(xml_df,
                          stubnames=stubnames,
                          i=['AwardID'],
                          j='InvestigatorNumber',
                          sep='',
                          suffix='\d+').reset_index()
```

```python
print("Transformed DataFrame:")
long_df.head()
```
